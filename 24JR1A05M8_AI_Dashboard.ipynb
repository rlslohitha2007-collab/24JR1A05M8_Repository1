```python
df.drop_duplicates(inplace=True)

df.fillna("Unknown", inplace=True)

print(df.isnull().sum())
```

```python
df.info()
```

```python
total_titles = len(df)

movies = len(df[df['type'] == 'Movie'])

tvshows = len(df[df['type'] == 'TV Show'])

countries = df['country'].nunique()

print("Total Titles:", total_titles)
print("Movies:", movies)
print("TV Shows:", tvshows)
print("Countries:", countries)
```

```python
import plotly.express as px
```

```python
fig = px.pie(
    df,
    names='type',
    title='Movies vs TV Shows'
)

fig.show()
```

```python
country_count = df['country'].value_counts().head(10)

fig = px.bar(
    x=country_count.index,
    y=country_count.values,
    title='Top 10 Countries'
)

fig.show()
```

```python
release = df['release_year'].value_counts().sort_index()

fig = px.line(
    x=release.index,
    y=release.values,
    title='Netflix Content Release Trend'
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x='release_year',
    y='duration_num',
    color='type',
    title='Release Year vs Duration'
)

fig.show()
```

```python
release = df['release_year'].value_counts().sort_index()

area_df = pd.DataFrame({
    'Year': release.index,
    'Count': release.values
})

fig = px.area(
    area_df,
    x='Year',
    y='Count',
    title='Netflix Content Growth Over Time'
)

fig.show()
```
